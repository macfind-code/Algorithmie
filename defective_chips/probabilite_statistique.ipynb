# Loi Binominale

```python
import pandas as pd

df2 = pd.read_csv('Dataset_lot_puce_defectueuses.csv')
```

```python
print(df2.head())
```

```python
df2.count()
```

```python
df2['défectueux'] = df2['défectueux'].apply(lambda x: 1 if x == 'Défectueux' else 0)
print(df2.head())
```

```python
proportion_defectueuses = df2['défectueux'].mean()
print(f'Proportion moyenne de puces défectueuses : {proportion_defectueuses}')
```

```python
defectueuses_par_lot = df2.groupby('lot_id')['défectueux'].sum()

import matplotlib.pyplot as plt
plt.hist(defectueuses_par_lot, bins=range(0, 21), edgecolor='black')
plt.title('Distribution du nombre de puces défectueuses par lot')
plt.xlabel('Nombre de puces défectueuses')
plt.ylabel('Nombre de lots')
plt.grid(True)
plt.show()
```

```python
# Probabilité qu'un lot ait au moins 5 puces défectueuses
probabilite_5_defectueuses = (defectueuses_par_lot >= 5).mean()
print(f'Probabilité qu\'un lot ait au moins 5 puces défectueuses : {probabilite_5_defectueuses}')
```

```python
# Pourcentage des lots rejetés (contenant 5 puces défectueuses ou plus)
pourcentage_rejet = (defectueuses_par_lot >= 5).mean() * 100
print(f'Pourcentage des lots rejetés : {pourcentage_rejet}%')
```
