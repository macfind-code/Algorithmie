```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('populations_lapins_renards.csv', parse_dates=['date'])

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

plt.plot(df['date'], df['lapin'], label='Rabbits', color='blue')
plt.plot(df['date'], df['renard'], label='Foxes', color='red')

plt.title('Rabbit and Fox Population Dynamics', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('populations_lapins_renards.csv', parse_dates=['date'])

rabbit_stats = {
    'initial_population': df['lapin'].iloc[0],
    'final_population': df['lapin'].iloc[-1],
    'max_population': df['lapin'].max(),
    'min_population': df['lapin'].min(),
    'total_change': df['lapin'].iloc[-1] - df['lapin'].iloc[0],
    'percent_change': ((df['lapin'].iloc[-1] - df['lapin'].iloc[0]) / df['lapin'].iloc[0]) * 100
}

fox_stats = {
    'initial_population': df['renard'].iloc[0],
    'final_population': df['renard'].iloc[-1],
    'max_population': df['renard'].max(),
    'min_population': df['renard'].min(),
    'total_change': df['renard'].iloc[-1] - df['renard'].iloc[0],
    'percent_change': ((df['renard'].iloc[-1] - df['renard'].iloc[0]) / df['renard'].iloc[0]) * 100
}

correlation = df['lapin'].corr(df['renard'])

print("Rabbit Population Statistics:")
for key, value in rabbit_stats.items():
    print(f"{key}: {value}")

print("\nFox Population Statistics:")
for key, value in fox_stats.items():
    print(f"{key}: {value}")

print(f"\nCorrelation between Rabbit and Fox Populations: {correlation:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('populations_lapins_renards.csv', parse_dates=['date'])

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

plt.scatter(df['lapin'], df['renard'], c=range(len(df)), cmap='viridis', alpha=0.7)
plt.colorbar(label='Time Progression')

plt.title('Phase Plane: Rabbits vs Foxes', fontsize=16)
plt.xlabel('Rabbit Population', fontsize=12)
plt.ylabel('Fox Population', fontsize=12)

plt.tight_layout()
plt.show()
```
